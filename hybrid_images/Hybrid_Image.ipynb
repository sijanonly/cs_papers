```

import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```
#parameters
sigma1 = 20
sigma2 = 20
kernel_size = 7
```

```

def do_low_pass(image, sigma, kernel_size):
    """Supressing low frequency components from the image"""
    return cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)
  
def do_high_pass(image, sigma, kernel_size):
    """Supressing low frequency components from the image"""
    return image - do_low_pass(image, sigma, kernel_size)

def prepare_hybrid_image(kernel_size, highpass_sigma, lowpass_sigma, highpass_img, lowpass_img):
    highfreq_image = do_high_pass(highpass_img, highpass_sigma, kernel_size)
    lowfreq_image = do_low_pass(lowpass_img, lowpass_sigma, kernel_size)

    return highfreq_image + lowfreq_image

```

```
cat = cv2.imread('cat.bmp', 1)
dog = cv2.imread('dog.bmp', 1)

plot_image = np.concatenate((cat, dog), axis=1)
plt.imshow(plot_image)
plt.show()
```

```
hybrid_image = prepare_hybrid_image(kernel_size, sigma1, sigma2, cat, dog)
plt.imshow(hybrid_image)
plt.show()
```

```
bird = cv2.imread('bird.bmp', 1)
plane = cv2.imread('plane.bmp', 1)

plot_image = np.concatenate((bird, plane), axis=1)
plt.imshow(plot_image)
plt.show()
```

```
hybrid_image = prepare_hybrid_image(kernel_size, sigma1, sigma2, bird, plane)
plt.imshow(hybrid_image)
plt.show()
```

```
einstein = cv2.imread('einstein.bmp', 1)
marilyn = cv2.imread('marilyn.bmp', 1)

plot_image = np.concatenate((einstein, marilyn), axis=1)
plt.imshow(plot_image)
plt.show()
```

```
hybrid_image = prepare_hybrid_image(kernel_size, sigma1, sigma2, einstein, marilyn)
plt.imshow(hybrid_image)
plt.show()
```
